```python
import geopandas as gpd
```

```python
 import folium
```

```python
from folium.plugins import HeatMap
```

```python
data = pd.read_csv(r"/Users/mahesh/Downloads/Sample Crime Dataset - Sheet1.csv")
print(data.head())
```

```python
# Convert the Date column to datetime format
data['Date'] = pd.to_datetime(data['Date'])
```

```python
# Ensure that the 'Latitude' and 'Longitude' columns are correctly named
data = data.rename(columns={'Lattitude': 'Latitude'})
```

```python
# Remove rows with missing values in 'Latitude' and 'Longitude'
data = data.dropna(subset=['Latitude', 'Longitude'])
```

```python
# Create a GeoDataFrame
gdf = gpd.GeoDataFrame(data, geometry=gpd.points_from_xy(data.Longitude, data.Latitude))
```

```python
# Plot a simple map of crime locations using folium
m = folium.Map(location=[41.8781, -87.6298], zoom_start=11)
```

```python
# Add crime data to the map
for idx, row in gdf.iterrows():
    folium.CircleMarker([row['Latitude'], row['Longitude']],
                        radius=1,
                        color='red',
                        fill=True).add_to(m)
```

```python
# Save map to an HTML file
m.save('crime_map.html')
```

```python
# Create a heatmap
m_heatmap = folium.Map(location=[41.8781, -87.6298], zoom_start=11)
```

```python
# Prepare data for the heatmap
heat_data = [[row['Latitude'], row['Longitude']] for index, row in data.iterrows()]
```

```python
# Add the heatmap layer
HeatMap(heat_data).add_to(m_heatmap)
```

```python
# Save heatmap to an HTML file
m_heatmap.save('crime_heatmap.html')
```

```python
# Crime type analysis
crimes_by_type = data['Primary Type'].value_counts()

# Plot the top 10 most common crime types
plt.figure(figsize=(12, 8))
top_crimes = crimes_by_type.head(10)
sns.barplot(x=top_crimes.values, y=top_crimes.index, palette='viridis')
plt.title('Top 10 Most Common Crime Types')
plt.xlabel('Number of Crimes')
plt.ylabel('Crime Type')
plt.show()
```

```python
# Crimes by year and type
crimes_by_year_type = data.groupby(['Year', 'Primary Type']).size().unstack()

# Plot the top 5 crime types by year
plt.figure(figsize=(15, 10))
top_crimes_by_year = crimes_by_year_type[top_crimes.index]
top_crimes_by_year.plot(kind='bar', stacked=True, colormap='viridis', figsize=(15, 10))
plt.title('Top 5 Crime Types by Year')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.legend(title='Crime Type')
```

```python
# Crimes by location description
crimes_by_location = data['Location Description'].value_counts()

# Plot the top 10 most common locations
plt.figure(figsize=(12, 8))
top_locations = crimes_by_location.head(10)
sns.barplot(x=top_locations.values, y=top_locations.index, palette='magma')
plt.title('Top 10 Crime Locations')
plt.xlabel('Number of Crimes')
plt.ylabel('Location Description')
plt.show()
```

```python
# Identify repeat case numbers
repeat_offenders = data['Case Number'].value_counts()
repeat_offenders = repeat_offenders[repeat_offenders > 1]
```

```python
# Display repeat offenders
print(repeat_offenders)
```

```python
# Filter data for repeat offenders
repeat_offender_data = data[data['Case Number'].isin(repeat_offenders.index)]
```

```python
# Sort data by case number and date
repeat_offender_data = repeat_offender_data.sort_values(by=['Case Number', 'Date'])
```

```python
# Calculate time between offenses for repeat offenders
repeat_offender_data['Time Between Offenses'] = repeat_offender_data.groupby('Case Number')['Date'].diff().dt.days
```

```python
# Display the first few rows of repeat offender data
print(repeat_offender_data.head())
```

```python
# Visualization

# Plot the number of repeat offenses
plt.figure(figsize=(12, 8))
sns.histplot(repeat_offenders.values, bins=20, kde=True)
plt.title('Number of Repeat Offenses')
plt.xlabel('Number of Offenses')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the time between offenses for repeat offenders
plt.figure(figsize=(12, 8))
sns.histplot(repeat_offender_data['Time Between Offenses'].dropna(), bins=20, kde=True)
plt.title('Time Between Offenses for Repeat Offenders')
plt.xlabel('Days Between Offenses')
plt.ylabel('Frequency')
plt.show()
```
